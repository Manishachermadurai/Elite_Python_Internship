# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('summer.csv')
df.head(10)
```

```python
df.isnull().sum()
```

### 1. In how many cities Summer Olympics is held so far?

```python
#Number of cities where olympics held
number_of_cities=len(df['City'].unique().tolist())
print("Number of cities where Olympics held is:{}".format(number_of_cities))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
#sport having most number of gold medals
gold_df = df[df["Medal"]=="Gold"]
gold_df.sort_values("Sport", ascending=False)
```

```python
top_5 =  pd.DataFrame(gold_df["Sport"].value_counts())
top_5 = top_5.reset_index()
top_5.columns = ["Sport", "Count"]
top_5
```

```python

sns.barplot(data = top_5[:5], x="Sport" ,y ="Count")
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
top_medal = df[["Sport", "Medal"]].groupby(["Sport"]).count().sort_values("Medal", ascending=False)
top_medal = top_medal.reset_index()
top_medal.columns = ['Sport', "Medal"]
top_medal
```

```python
sns.barplot(data=top_medal.head(), x="Sport", y="Medal")
```

### 4. Which player has won most number of medals? (Top 5)

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
gold_df = pd.DataFrame(gold_df["Athlete"].value_counts())
gold_df = gold_df.reset_index()
gold_df.columns = ["Athlete", "Medal"]
gold_df
```

```python
ax = sns.barplot(data = gold_df.head(), x="Athlete", y="Medal")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
#India's first gold medal
india_first_medal = df[(df["Country"]=="IND") & (df["Medal"]=="Gold")]
print("India won the first medal in {}.".format(india_first_medal["Year"].iloc[1]))
```
